```python
# In Python functions can take functions as arguments and invoke those functions, lets see an example

def is_even(x):
    return x % 2 == 0

def only_even_numbers(list, is_even):
    result = []
    for x in list:
        if(is_even(x)):
            result.append(x)
    return result
            
only_even_numbers([1, 2, 3, 4, 5], is_even)
```

```python
# Other really helpful functions to work with lists filter, map, reduce
```

```python
# filtering elements in a list using built-in filter method
numbers = [1, 2, 3, 4, 5]
```

```python
def is_even_number(num):
    return num % 2 == 0

filtered_list = filter(is_even_number, numbers)
print(filtered_list)
```

```python
# creating a new list from existing list using built-in map method
def square_it(num):
    return num * num
squared_list = map(square_it, numbers)
print(squared_list)
```

```python
# So we have now learnt about Dictionary or Map in Python, now lets learn about Lambda functions
# a lambda function, is a normal function but without any name
# Lambdas create a shorthand to define functions and store them in variables
add_function = lambda x, y: x + y
add_function(3, -4)
```

```python
# now that we can have function variables, we can pass functions to other functions as arguments
def filter_list(list, filter_function):
    result = []
    for x in list:
        if(filter_function(x)):
            result.append(x)
    return result
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train=pd.read_csv("height-weight.csv", sep='\t')
```

```python
train.head()
```

```python
train.info()
```

```python
train.describe()
```

```python
train['height'] = train['Height(Inches)'].map(lambda inches: round(inches * 0.0833333, 2))
train.head()
```

```python
train['weight'] = train['Weight(Pounds)'].map(lambda pounds: round(pounds * 0.453592, 2))
train.head()
```

```python
df = train[['height', 'weight']]
df.head()
```

```python
len(df.index)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(df['height'], df['weight'])
plt.title('Height Weight Scatter Plot')
plt.xlabel('Height (in feet)')
plt.ylabel('Weight (in kgs)')
plt.show()
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(df['height'], df['weight'])
```

```python
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(df['height'])

plt.scatter(df['height'], df['weight'])
plt.plot(df['height'], fitLine, c='r')
plt.show()
```

```python
from sklearn import tree

input_file = "./resume_shortlist.csv"
df = pd.read_csv(input_file, header = 0)

df.head()
```

```python
d = {'Y': 1, 'N': 0}
df['Shortlisted'] = df['Shortlisted'].map(d)
df['Referral'] = df['Referral'].map(d)

d = {'BTech': 0, 'MTech': 1}
df['Degree'] = df['Degree'].map(d)

df.head()
```

```python
features = df.columns[:4]
features
```

```python
y = df["Shortlisted"]
X = df[features]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
print(clf.predict([[1, 90, 3, 1]]))
```

```python
print(clf.predict([[0, 30, 5, 0]]))
```
